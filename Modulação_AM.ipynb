```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
f_m = 0.05 # Frequência da onda moduladora em Hz
f_c = 0.8 # Frequência da onda portadora em Hz
t = np.linspace(0, 10/f_m, 10000) # Vetor de tempo do sinal

# Sinal da onda moduladora
s_m = np.sin(2 * np.pi * f_m * t)

# Sinal da onda portadora
s_p = np.cos(2 * np.pi * f_c * t)

# Sinal DSB-SC
s_t = s_m * s_p

# Plotagem  do sinal DSB-SC
plt.figure(figsize=(12, 4))
plt.plot(t, s_t, 'g', label='Sinal DSB-SC')
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.title('Onda DSB_SC Modulada por uma Onda Senoidal de 0.05 Hz')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotagem  dos sinais para comparação
plt.figure(figsize=(12, 4))
plt.subplot(3,1,1)
plt.plot(t, s_t, 'g', label='Sinal DSB-SC')
plt.title('Onda DSB_SC Modulada por uma Onda Senoidal de 0.05 Hz')
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(12, 4))
plt.subplot(3,1,2)
plt.plot(t, s_m, 'y', label='Sinal da Moduladora')
plt.title('Onda Moduladora')
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(12, 4))
plt.subplot(3,1,3)
plt.plot(t, s_p, label='Sinal da Portadora')
plt.title('Onda Portadora')
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Parâmetros
f_m = 0.05 # Frequência da onda moduladora em Hz
f_c = 0.8 # Frequência da onda portadora em Hz
t = np.linspace(0, 10/f_m, 10000) # Vetor de tempo do sinal

# Sinal da onda moduladora
s_m = np.sin(2 * np.pi * f_m * t)

# Sinal da onda portadora
s_p = np.cos(2 * np.pi * f_c * t)

# Sinal DSB-SC
s_t = s_m * s_p

# Espectro de amplitude utilizando FFT
N = len(t)
T = t[1] - t[0] # Intervalo de amostragem
yf = np.fft.fft(s_t)
xf = np.linspace(-1.0/(2.0*T), 1.0/(2.0*T), N) # Intervalo de frequência

# Plotagem do espectro de amplitude
plt.figure(figsize=(12, 4))
plt.plot(xf, np.fft.fftshift(2.0/N * np.abs(yf)), label='Espectro de Amplitude')
plt.xlabel('Frequência(Hz)')
plt.ylabel('Amplitude')
plt.title('Espectro de Amplitude da Onda Modulada DSB-SC')
plt.grid(True)
plt.legend()
plt.show()
```
